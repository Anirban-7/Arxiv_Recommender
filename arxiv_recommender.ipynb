```python
import pandas as pd
import numpy as np
from sentence_transformers import SentenceTransformer, util
import regex as re
```

```python
#Loading the data set
df = pd.read_csv('./data/df_experiment').drop('Unnamed: 0', axis=1)
```

```python
#printing the first five rows of the data set:
df.columns
```

```python
#calling the model object
model = SentenceTransformer('all-MiniLM-L6-v2')
```

```python
#Encoding the summaries of the articles
summaries = df.summary.to_list()
embeddings = model.encode(summaries)
```

```python
#store sentences and embeddings
import pickle

with open('embeddings.pkl',"wb") as fOut:
  pickle.dump({'Summary':df['summary'],'Embeddings': embeddings},fOut,protocol=pickle.HIGHEST_PROTOCOL)
```

```python
# function to clean the input article
def cleaning(new_article):
    new_article = new_article.lower()
    cleaned = re.sub('[^A-Za-z0-9]+',' ',new_article)
    return cleaned
#print(cleaning(trial_article))
```

```python
def recommendation(new_article):
    #cleaning the abstract
    cleaned = cleaning(new_article)
    #embedding for the new article
    input_embedding = model.encode(cleaned)
    similarity_score=[]
    #storing the cosine smilarity score with the exisiting papers
    for embedding in embeddings:
        similarity_score.append(util.cos_sim(input_embedding,embedding).item())
    reco_idx = np.array(similarity_score).argmax()
    print('Our Top Recommendation')
    print('Title: '+str(df['title'][reco_idx]))
    print('Authors: '+str(df['authors'][reco_idx]))
    print('Abstract: '+str(df['summary'][reco_idx]))
   # print('You can refer to: '+str(df['summary'][reco_idx]))
```

```python
if __name__=='__main__':
    input_article = input('Please enter the abstract summary: ')
    recommendation(input_article)
```
